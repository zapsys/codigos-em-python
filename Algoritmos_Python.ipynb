```python
# Calculadora simples para dois números (inteiros ou decimais)
n1 = float(input('Informe o primeiro número: '))
n2 = float(input('Informe o segundo número: '))
op = input('Informe a operação desejada: ')


# Estrutura condicional 'if/elif' e 'while' para avaliar se o segundo número é
# zero na divisão. Se for 0 solicita nova entrada

if op == '+':
  print('Soma: ', n1 + n2)
elif op == '-':
  print('Subtração: ', n1 - n2)
elif op == '*':
  print('Multiplicação: ', n1 * n2)
elif op == '/':
  while n2 == 0:
    print('\nNão é possível dividir por zero! Informe outro número.\n')
    n2 = float(input('Informe o segundo número: '))
  print('Divisão: ', n1 / n2)
elif op == '%':
  print('Resto da operação: ', n1 % n2)
```

```python
# Algoritmo para soma n números inteiros

q = int(input('Informe a quantidade de números para somar: '))

# Inicialização da variável de controle
i = 1

# Inicialização da variável soma
soma = 0

# Laço de repetição
while i <= q:
  n = int(input(f'Informe o {i}° número: ')) # Solicita a entrada do número
  i = i + 1       # Incremento da variável de controle
  soma = soma + n # Soma o valor da soma anterior com o valor de 'n' informado

print(f'A soma dos {q} números é igual a: ', soma) # Imprime a soma dos 'q' números
```

```python
# Progressão aritmética com passo 1
# Exemplo: Soma dos números de 1 a 100 -> 1 + 2 + 3 +...+ 100

# Valor inicial da progressão
ai = int(input('Informe o valor inicial da progressão: '))

# Quantidade de termos ou posição do enésimo termo
n = int(input('Informe a posição do enésimo termo: '))

# Valor do enésimo termo da progressão
an = int(input('Informe o valor do enésimo termo: '))

# Soma dos termos (Fórmula de Gauss)
Sn = ((ai + an)*n)/2

print(f'Soma dos valores de {ai} a {an}: ', Sn)
```

```python
# Algoritmo para verificar se um número é primo ou não e mostrar seus divisores
# Um número primo é divisível por 1 e ele mesmo, ou seja, 2 divisores

n = int(input('Informe um número inteiro: '))

# Variável para contar o divisores
count = 0

# Variável de controle
i = 1
print('----------------------')
print(f'Divisores de {n}: ')
print('----------------------')
# Laço de repetição
while i <= n:
  if (n % i) == 0:
    count = count + 1
    div = int(n/i)  # Variável para armazenar os divisores
    print(div)      # Imprime os divisores
  i = i + 1         # Incrementa a variável de controle
print('----------------------')
if count < 2 or count > 2:
  print(f'O número {n} não é primo')
else:
  print(f'O número {n} é primo')
```

```python
# Algoritmo para fazer o controle de gastos mensal

# Quantidade de gastos do mês
q = int(input('Quantidade de gastos desse mês: '))

# Variável para a soma dos gastos
soma = 0

# Usamos a função range para definir o limite dos gastos. De 1 a q (< q+1)
print('---------------------------------')
for i in range(1,q+1):
  gasto = float(input(f'Despesa {i}: '))
  soma = soma + gasto
print('---------------------------------')
print(f'Gastos do mês -> R$ {soma}')
```

```python
# Contando caracteres em um texto e contagem de palavras

texto = input('Informe uma palavra ou texto: ')

# O método split() remove o delimitador definido, no caso o espaço ''
# O método join retorna um novo texto é 'juntado' (.join) com o
# restante da string

n_texto = ''.join(texto.split())

count = 0
count_c = 0

# A variável 't' percorre todos os caracteres do texto (espaços inclusive)
for t in texto:
  count = count + 1
# A variável 'c' percorre todos os caracteres do texto (sem espaços)
for c in n_texto:
  count_c = count_c + 1
print('Quantidade de palavras: ', count - (count_c - 1))
print('Quantidade de caracteres: ', count_c)
```
